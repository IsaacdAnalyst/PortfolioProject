**USING THE MOVING AVERAGE STRATEGY(MA)**

I will be building a simple predictor using the moving average stategy.
The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks, or any time period the trader chooses.A moving average helps cut down the amount of noise on a price chart. Look at the direction of the moving average to get a basic idea of which way the price is moving.

I will be using the 10 days and 50 days moving average.

**CACULATE THE PROBAILITY OF THE STOCK PRICE WILL DROP OVER A CERTAIN PERCENTAGE IN A DAY**

```python
# probability that the stock price will drop over 5% in a day
prob_return1 = norm.cdf(-0.05, mu, sigma)
print('The Probability to drop over 5% is ', prob_return1)

# probability that the stock price will drop over 5% in a day
prob_return1 = norm.cdf(-0.10, mu, sigma)
print('\nThe Probability to drop over 10% is ', prob_return1)
```

**CALCULATE THE PROBABILTY OF THE STOCK PRICE WILL DROP OVER A CERTAIN PERCENTAGE IN A YEAR**

```python
# drop over 40% in 220 days
mu220 = 220*mu                          # we have 220 trading days in a year.
sigma220 = (220**0.5) * sigma
print('The probability of dropping over 40% in 220 days is ', norm.cdf(-0.4, mu220, sigma220))

#drop over 20% in 220 days
mu220 = 220*mu                          # we have 220 trading days in a year.
sigma220 = (220**0.5) * sigma
print('\nThe probability of dropping over 20% in 220 days is ', norm.cdf(-0.2, mu220, sigma220))
```
